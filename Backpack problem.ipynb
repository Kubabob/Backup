```python
import pygad
import numpy as np
import random
```

```python
test = np.array([1,0,1])
test2 = [2,5,10]

print(np.sum(test*test2))
```

```python
prices = [100, 300, 200, 40, 500, 70, 100, 250, 300, 280, 300]
weights = [7, 7, 6, 2, 5, 6, 1, 3, 10, 3, 15]
```

```python
prices = [random.randint(10, 500) for _ in range(10)]
weights = [random.randint(1, 15) for _ in range(10)]
```

```python
def fitness_func(ga_instance, solution, solution_idx):
    weight = np.sum(solution*weights)
    if weight <= 25:
        money = np.sum(solution*prices)
        fitness = money/np.sum(prices)
    else:
        fitness = 0
    return fitness
    
```

```python
ga_instance = pygad.GA(
    gene_space=[0,1],
    num_generations=10,
    num_parents_mating=5,
    fitness_func=fitness_func,
    sol_per_pop=10,
    num_genes=len(prices),
    parent_selection_type='sss',
    keep_parents=2,
    crossover_type='single_point',
    mutation_type='random'
)
```

```python
ga_instance.run()

solution, solution_fitness, solution_index = ga_instance.best_solution()
print(f'Prices: {prices}')
print(f'Weights: {weights}')
print(f'Best solution: {solution*prices}')
print(f'Solution fitness: {solution_fitness}')
print(f'Money: {np.sum(solution*prices)}')
print(f'Weight: {np.sum(solution*weights)}')

ga_instance.plot_fitness()
```
